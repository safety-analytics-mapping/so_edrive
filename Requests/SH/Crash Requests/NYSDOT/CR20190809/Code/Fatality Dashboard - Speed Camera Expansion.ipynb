### ALL 2019

```python
mon_fri_6am_10pm_1 = db2.query_to_table(cdb, """

                        select  sum(case when pos= 'BI' then 1 else 0 end) as BI,
                                sum(case when pos= 'DR' then 1 else 0 end) as DR,
                                sum(case when pos= 'MO' then 1 else 0 end) as MO,
                                sum(case when pos= 'PD' then 1 else 0 end) as PD,
                                sum(case when pos= 'PS' then 1 else 0 end) as PS
                        from fatality_nycdot_current 
                        where yr = 2019 
                        and dow in (2,3,4,5,6)
                        and left(atime,2)::int between 6 and 22
                        
                        """)

#mon_fri_6am_10pm_1

mon_fri_inverse_1 = db2.query_to_table(cdb, """

                        select  sum(case when pos= 'BI' then 1 else 0 end) as BI,
                                sum(case when pos= 'DR' then 1 else 0 end) as DR,
                                sum(case when pos= 'MO' then 1 else 0 end) as MO,
                                sum(case when pos= 'PD' then 1 else 0 end) as PD,
                                sum(case when pos= 'PS' then 1 else 0 end) as PS
                        from fatality_nycdot_current 
                        where yr = 2019 
                        and dow in (2,3,4,5,6)
                        and (left(atime,2)::int<6 or left(atime,2)::int>22)
                        
                        """)

#mon_fri_inverse_1

sun_sat_6am_10pm_1 = db2.query_to_table(cdb, """

                        select  sum(case when pos= 'BI' then 1 else 0 end) as BI,
                                sum(case when pos= 'DR' then 1 else 0 end) as DR,
                                sum(case when pos= 'MO' then 1 else 0 end) as MO,
                                sum(case when pos= 'PD' then 1 else 0 end) as PD,
                                sum(case when pos= 'PS' then 1 else 0 end) as PS
                        from fatality_nycdot_current 
                        where yr = 2019 
                        and dow not in (2,3,4,5,6)
                        and left(atime,2)::int between 6 and 22

                        
                        """)

#sun_sat_6am_10pm_1

sun_sat_inverse_1 = db2.query_to_table(cdb, """

                        select  sum(case when pos= 'BI' then 1 else 0 end) as BI,
                                sum(case when pos= 'DR' then 1 else 0 end) as DR,
                                sum(case when pos= 'MO' then 1 else 0 end) as MO,
                                sum(case when pos= 'PD' then 1 else 0 end) as PD,
                                sum(case when pos= 'PS' then 1 else 0 end) as PS
                        from fatality_nycdot_current 
                        where yr = 2019 
                        and dow not in (2,3,4,5,6)
                        and (left(atime,2)::int<6 or left(atime,2)::int>22)
                        
                        """)

#sun_sat_inverse_1
```

```python
dash = pd.DataFrame()

dash = dash.append(mon_fri_6am_10pm_1, ignore_index=True)
dash = dash.append(mon_fri_inverse_1, ignore_index=True)
dash = dash.append(sun_sat_6am_10pm_1, ignore_index=True)
dash = dash.append(sun_sat_inverse_1, ignore_index=True)
```
