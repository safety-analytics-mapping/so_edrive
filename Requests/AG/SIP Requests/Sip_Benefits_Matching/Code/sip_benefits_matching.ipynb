### Bike Lane

```python
sip_treatments_bl = db.dfquery("""SELECT DISTINCT qid, full_treatment, full_treatment_int
                                   FROM public.sip_flattened_tree
                                   WHERE full_treatment like '%New Facility%'
                                   AND (full_treatment like '%Standard%' OR full_treatment like '%Curbside%' OR full_treatment like '%Buffered%')
                                   AND full_treatment not like '%Removed%'
                                   """)

sip_treatments_bl['improvement']= 'Bike Lane'
sip_treatments_bl['benefit'] = sip_benefits.loc[sip_benefits.Improvement=='Bike Lane'].reset_index()['Benefit 1']

sip_treatments_bl['benefit'] = sip_treatments_bl['benefit'].ffill(axis=0)
```
