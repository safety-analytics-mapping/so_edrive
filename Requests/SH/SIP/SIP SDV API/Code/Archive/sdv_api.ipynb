```python
def grabSipProjects(db,year):
 
    """
    Fucntion to grab SIP crash history
    :param year (int): year of SIP crash history data desired
    """
    

    sip = db.dfquery("""
                          SELECT pid 
                          FROM sip_projects
                          WHERE sip_year = {yr}
                          AND status = '15'
                          """.format(yr=year))  
    
    return map(int,list(sip.pid))

def GetSIPCrashHistory(pid):

    """
    Fucntion to grab SIP crash history
    :param pid: SIP project id
    """
    
    #1 access the SIP crash history
    
    _req_id = 0;
    _url = _api_url + "Crash/GetCrashHistory";
    _data = {
            "req_id":0,
            "sip_id": "%s" % pid,
            "use_sip_geometry": "Y",
            "prj_title": "Crash History Sample",
            "from": "01/01/2014",
            "to": "12/31/2018",
            "usr_id": "Sam",
            "crashfilters": []
        }
    
    #print _data    
    d = pd.DataFrame([_data])
    
    r = requests.post(_url, data=_data, headers=_api_hdr)
    print(r.status_code)

    json_response = r.json()
    #dictionary = json.dumps(r.json(), sort_keys = True, indent = 4)

    return (d,json_response)

def SIP_Subset(data_table):
    
    """
    Function to grab subset of data from data table
    :param cross_tab: cross_tab of data desired
    """

    data = pd.DataFrame(data_table[1]["CrashByAgeGroup"])
    SeniorInjuries = data.loc[data['AgeGroup'] == 'Senior(65-120)'][['PercentKnown','PercentKnownBoro']]
    SI = SeniorInjuries.reset_index(drop=True)
    
    result = pd.concat([data_table[0], SI], axis=1)
    
    return result[['sip_id','PercentKnown','PercentKnownBoro']]

def run(db):

    sip_projects = grabSipProjects(db,2019)

    df = pd.DataFrame()
    
    for i in sip_projects[:5]:

        data_table = GetSIPCrashHistory(i)
        df = df.append(SIP_Subset(data_table))
        
    return df   
```
